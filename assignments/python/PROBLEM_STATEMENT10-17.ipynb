```python
update_dict1 = {'Daman & Diu':'Daman and Diu', 'Dadra & Nagar Haveli':'Dadra and Nagar Haveli','Orissa':'Odisha',
              'Puducherry':'Pondicherry','Jammu & Kashmir':'Jammu and Kashmir','Andaman & Nicobar Islands':'Andaman and Nicobar Islands',
              'Delhi':'Delhi'}

df_grouped['State/UT'] = df_grouped['State/UT'].apply(lambda x : update_dict1[x] if x in update_dict1.keys() else x)

df_grouped.set_index('State/UT',inplace=True)
df_grouped
```

```python
all_india = df_grouped['Population'].sum()
df_grouped.loc['All India'] = all_india
df_grouped
```

```python
pd.set_option('display.float_format', lambda x: '%.1f' % x)
df_grouped = pd.DataFrame(round(df_grouped['Population'],1))
df_grouped
```

```python
merged_df = pd.merge(hos_df, df_grouped, on=['State/UT'])
```

```python
merged_df
```

```python
merged_df['Hospitalbeds'] = pd.to_numeric(merged_df['HospitalBeds'])
```

```python
merged_df['Beds_per_10k'] = (merged_df['Hospitalbeds']/merged_df['Population'])*10000
```

```python
national_average = (merged_df["Hospitalbeds"].sum() / (merged_df["Population"].sum())) * 10000
```

```python
plt.bar(merged_df["State/UT"], merged_df["Beds_per_10k"], color="pink",label='States')
plt.axhline(national_average, color="red", linestyle="--", label="National average")
plt.xlabel("Number of hospital beds per 10,000 people")
plt.xticks(rotation=90)
plt.ylabel("State/UT")
plt.title("Hospital beds per 10,000 people in each State/UT in India")
plt.legend()
plt.show()
```

```python
total_hos_beds
```

```python
total_hos_beds['Average_beds'] = total_hos_beds['Total_Beds']/total_hos_beds['Total_Hospitals']
total_hos_beds
```

```python
total_hos_beds['Expected_total_beds']=(total_hos_beds['Population']*total_hos_beds['Gap'])/1000
total_hos_beds
```

```python
total_hos_beds['Hospitals_needed']=total_hos_beds['Expected_total_beds']/total_hos_beds['Average_beds']
total_hos_beds
```

```python
hos_needed = total_hos_beds[['State/UT','Gap','Expected_total_beds','Hospitals_needed']]
hos_needed
```
