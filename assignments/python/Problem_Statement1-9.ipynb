```python
def capital(st):
    st=st.split()
    for i in range(0,len(st)):
        if st[i] != 'AND':
            st[i] = st[i].capitalize()
        elif st[i] == 'AND':
            st[i] = st[i].lower()
    st = ' '.join(st)
    return st

```

```python
df1['State/UT']= df1['State/UT'].apply(lambda x:capital(x))
```

```python
# Rural_Households,Urban_Households,
#Households,Condition_of_occupied_census_houses_Dilapidated_Households

#Total Number of households
#Total Number of Livable
#Total Number of Dilapidated

merge_copy = merged_data[['State name','District Name','Rural_Households', 'Urban_Households',
                         'Households','Rural/Urban','Total Number of households','Total Number of Livable',
                          'Total Number of Dilapidated','Latrine_premise']].copy()
```

```python
def replace_households(test):
    if test['Rural/Urban']=='Rural':
        test['Total Number of households']=test['Rural_Households']
    elif test['Rural/Urban']=='Total':
        test['Total Number of households']=test['Households']
    else:
        test['Total Number of households']=test['Urban_Households']
    return test['Total Number of households'] 
    
    
def replace_livable(test):
    if test['Rural/Urban']=='Rural':
        test['Total Number of Livable']=test['Total Number of Livable']*test['Rural_Households']/100
    elif test['Rural/Urban']=='Total':
        test['Total Number of Livable']=test['Total Number of Livable']*test['Households']/100
    else:
        test['Total Number of Livable']=test['Total Number of Livable']*test['Urban_Households']/100
    return test['Total Number of Livable']

    
def replace_dilapidated(test):
    if test['Rural/Urban']=='Rural':
        test['Total Number of Dilapidated']=test['Total Number of Dilapidated']*test['Rural_Households']/100
    elif test['Rural/Urban']=='Total':
        test['Total Number of Dilapidated']=test['Total Number of Dilapidated']*test['Households']/100 
    else:
        test['Total Number of Dilapidated']=test['Total Number of Dilapidated']*test['Urban_Households']/100 
        
    return test['Total Number of Dilapidated']
        
        
def replace_latrine(test):
    if test['Rural/Urban']=='Rural':
        test['Latrine_premise']=test['Latrine_premise']*test['Rural_Households']/100
    elif test['Rural/Urban']=='Total':
        test['Latrine_premise']=test['Latrine_premise']*test['Households']/100 
    else:
        test['Latrine_premise']=test['Latrine_premise']*test['Urban_Households']/100
    return test['Latrine_premise']
        
```

```python
merge_copy['Total Number of households'] = merge_copy.apply(lambda x:replace_households(x),axis=1)
merge_copy['Total Number of Livable'] = merge_copy.apply(lambda x:replace_livable(x),axis=1)
merge_copy['Total Number of Dilapidated'] = merge_copy.apply(lambda x:replace_dilapidated(x),axis=1)
merge_copy['Latrine_premise'] = merge_copy.apply(lambda x:replace_latrine(x),axis=1)
```
